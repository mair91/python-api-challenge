```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

urlOW = "http://api.openweathermap.org/data/2.5/weather?"

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
citiesName = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in citiesName:
        citiesName.append(city)

# Print the city count to confirm sufficient count
len(citiesName)
```

```python
counter = 1
cityNameWeather = []
countryWeather = []
latWeather = []
lngWeather = []
maxTempWeather = []
humidWeather = []
cloudWeather = []
windWeather = []
dateWeather = []

for city in citiesName:
    query = urlOW + "q=" + city + "&units=metric" + "&appid=" + weather_api_key

    try:
        response = requests.get(query).json()
        cityNameWeather.append(response['name'])
        countryWeather.append(response['sys']['country'])
        latWeather.append(response['coord']['lat'])
        lngWeather.append(response['coord']['lon'])
        maxTempWeather.append(response['main']['temp_max'])
        humidWeather.append(response['main']['humidity'])
        cloudWeather.append(response['clouds']['all'])
        windWeather.append(response['wind']['speed'])
        dateWeather.append(response['dt'])

        print(f'Processing Record {counter} of {len(citiesName)} | {city}')

        counter += 1
    except:
        print('City not found. Skipping...')
```

```python
weather_df = pd.DataFrame(list(zip(cityNameWeather, countryWeather, latWeather, lngWeather, maxTempWeather, humidWeather, cloudWeather, windWeather, dateWeather))) 
weather_df.columns = ['City', 'Country', 'Latitude', 'Longitude', 'MaxTemp', 'Humidity', 'Clouds', 'WindSpeed', 'Date']
weather_df.head()
```

```python
weather_df.to_csv ('output/weather_dataframe.csv', index = False, header=True)
```

```python
def linearRegression(xValues,yValues, x_Label, y_Label, theTitle, xEq, yEq):
    x_values = xValues
    y_values = yValues
    (slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
    regress_values = x_values * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    plt.scatter(x_values,y_values)
    plt.plot(x_values,regress_values,"r-")
    plt.annotate(line_eq,(xEq,yEq),fontsize=15,color="red")
    plt.xlabel(x_Label)
    plt.ylabel(y_Label)
    plt.title(theTitle)
    print(f"The r-squared is: {rvalue**2}")
    plt.show()
```

```python
northern_df = weather_df.loc[weather_df['Latitude'] > 0]
southern_df = weather_df.loc[weather_df['Latitude'] < 0]
```
